```python
# Fibbonacci without using memoization

def feb(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1    
    else:
        return feb(n-1) + feb(n-2)
   
```

```python
feb(38) ## Takes Time to compute the result. It is an exponential time complexity O(2^n) 


```

```python
## FIBONACCI USING MEMOIZATION
def fibo(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 0:
        return 0
    elif n == 1:
        return 1    
    else:
        memo[n] = fibo(n-1, memo) + fibo(n-2, memo)
        return memo[n]
```

```python
fibo(48) ## Takes very less time to compute the result. It is a linear time complexity O(n) and space complexity O(n) due to the memoization dictionary.
```
