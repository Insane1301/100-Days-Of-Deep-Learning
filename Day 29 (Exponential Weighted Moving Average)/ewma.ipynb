```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DailyDelhiClimateTest.csv')
```

```python
df = df[['date', 'meantemp']]
df
```

```python
plt.scatter(df['date'], df['meantemp'],color='blue', label='meantemp')
plt.xlabel('Date')
plt.ylabel('Mean Temperature')
plt.title('Delhi Climate Data')
plt.legend()
plt.show()
```

```python
x1 = df['meantemp'].ewm(alpha=0.9).mean()
```

```python
x1
```

```python
df['ewma'] = x1
df
```

```python
x2 = df['meantemp'].ewm(alpha=0.1).mean()
```

```python
df['ewma'] = x2
df
```
